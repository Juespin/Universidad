# Ecuación de Bernoulli y Continuidad
El presente código gráfica las alturas en los distintos puntos de un equipo de Bernoulli, a la vez que calcula las cabezas de presión, velocidad y total para demostrar la conservación de esta última.

## Objetivos.
- Comprobar la ecuación de continuidad.
- Comprobar el teorema de Bernoulli.
- Comprobar la variación de la presión a medida que un fluido modifica su velocidad.
- Analizar los conceptos de cabeza de velocidad estática y total y su influencia en el flujo.

## Marco teórico.
La Ecuación de Bernoulli relaciona presión, velocidad y altura en los puntos situados a lo largo de una línea de corriente, de la siguiente forma:

$$ \frac{P_1}{\gamma} + \frac{V_{1}^2}{2g} + Z_1 = \frac{P_2}{\gamma} + \frac{V_{2}^2}{2g} + Z_2 $$

Donde:
  - $P$: Valor de presión.
  - $V$: Valor de velocidad.
  - $Z$: Valor de altura.
  - $g$: Valor de la gravedad.
  - $\gamma$: Peso específico del fluido.

Para el equipo de Bernoulli, las cabezas de presión hidrostáticas dadas por las columnas corresponden a $P = \gamma \times h$, además, para una posición horizontal del tubo se tiene que $Z_1 = Z_2$, por lo que se simplifica la ecuación a:

$$ h_1 + \frac{V_{1}^2}{2g} = h_2 + \frac{V_{2}^2}{2g} $$

Donde:
  - $h$: Altura de la columna de agua.
  - $V$: Valor de velocidad.
  - $g$: Valor de la gravedad.

Y además, $h$ representa la cabeza de presión,  $\frac{V^2}{2g}$ representa la de velocidad, por lo que la cabeza total se representa como:

$$ H = h + \frac{V^2}{2g} $$

Si no se conocen las velocidades pero si el caudal, se utiliza la ecuación de continuidad de la siguiente manera:

$$ Q = A \times V \therefore V = \frac{Q}{A} \Rightarrow V = \frac{Q}{\pi \left( \frac{D}{2} \right)^2 } = \frac{4 Q}{\pi D^2} \therefore V^2 = \frac{16 Q^2}{\pi^2 D^4} $$

Donde:
  - $Q$: Caudal.
  - $A$: Área.
  - $D$: Diámetro.

De manera que la cabeza total se calcula como:

$$ H = h + \frac{8Q^2}{\pi^2 g D^4} $$

```python
import matplotlib.pyplot as plt

data_1 = {'h': [138.3333, 118.3333,  91.6667,  70, 43.3333, 23.3333], 'Q': 119.4156, 't': 2.5}
data_2 = {'h': [273.3333, 253.3333, 228.3333, 203.3333, 173.3333, 138.3333], 'Q': 107.9021, 't': 2.3333}
data_3 = {'h': [151.6667, 128.3333, 103.3333, 80, 53.3333, 20], 'Q': 121.9231, 't': 2.3333}

manometer = ['f', 'e', 'd', 'c', 'b', 'a']

plt.figure(figsize=(10, 5))
plt.bar(manometer, data_2['h'], color='maroon', label=f"Caudal: {data_2['Q']} L/s")
plt.bar(manometer, data_3['h'], color='navy', label=f"Caudal: {data_3['Q']} L/s")
plt.bar(manometer, data_1['h'], color='green', label=f"Caudal: {data_1['Q']} L/s")
plt.bar(['a'], data_3['h'][-1], color="navy")
#plt.title('Alturas promedio')
plt.xlabel('Manómetros')
plt.ylabel('Altura [mm]')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datas = {
'data_1' : {'h': [138.3333, 118.3333,  91.6667,  70, 43.3333, 23.3333], 'Q': 119.4156, 't': 2.5}
,'data_2': {'h': [273.3333, 253.3333, 228.3333, 203.3333, 173.3333, 138.3333], 'Q': 107.9021, 't': 2.3333}
,'data_3': {'h': [151.6667, 128.3333, 103.3333, 80, 53.3333, 20], 'Q': 121.9231, 't': 2.3333}
}
diameter = [25, 14.6, 12.4, 11.3, 10.6, 10]
p = []
v = []
for data in datas:
  d = 0
  for h in datas[data]['h']:
    p.append(h*1E-3)
    v.append((8*(datas[data]['Q']*1E-6)**2)/((np.pi**2)*9.81*(diameter[d]*1E-3)**4))
    d += 1

p1 = pd.DataFrame(p[0:6])
v1 = pd.DataFrame(v[0:6])
p2 = pd.DataFrame(p[6:12])
v2 = pd.DataFrame(v[6:12])
p3 = pd.DataFrame(p[12:])
v3 = pd.DataFrame(v[12:])

print(p1)
print(v1)
print(p1 + v1)
print('\n')
print(p2)
print(v2)
print(p2 + v2)
print('\n')
print(p3)
print(v3)
print(p3 + v3)
```
