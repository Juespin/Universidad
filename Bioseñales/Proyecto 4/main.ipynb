```python

ruta_ecg = "ecg_data_denoised/ECGDataDenoised"
detector_ecg = Detectors(sampling_frequency=500)

def encontrar_r_peaks(ecg_signal):
    return detector_ecg.hamilton_detector(ecg_signal.to_numpy())

def obtener_nn_intervals(r_locs, fs=500):
    return np.diff(r_locs) * (1000 / fs)

# Listas para guardar resultados y seales descartadas
resultados_hrv = []
ecg_no_validos = []

# Procesamiento de los archivos
for idx, registro in tqdm(df.iterrows(), total=len(df), desc="Analizando registros ECG"):
    nombre_archivo = f"{registro['FileName']}.csv"
    ruta_archivo = os.path.join(ruta_ecg, nombre_archivo)

    if not os.path.isfile(ruta_archivo):
        print(f"No se encontr el archivo: {nombre_archivo}")
        continue

    try:
        datos_ecg = pd.read_csv(ruta_archivo, header=None)
        ecg_signal = datos_ecg.iloc[:, 1]
        r_peaks = encontrar_r_peaks(ecg_signal)
        intervalos_nn = obtener_nn_intervals(r_peaks)

        if len(intervalos_nn) > 2:
            caracteristicas = get_time_domain_features(intervalos_nn.tolist())
            caracteristicas["Cantidad_R_Peaks"] = len(r_peaks)
            caracteristicas["Registro"] = registro["FileName"]
            caracteristicas["Estado"] = registro["Rhythm"]
            resultados_hrv.append(caracteristicas)
        else:
            ecg_no_validos.append((ecg_signal, registro["FileName"]))
    except Exception as error:
        print(f"Fallo al procesar {nombre_archivo}: {error}")

# Guardar resultados en un CSV
df_resultados_hrv = pd.DataFrame(resultados_hrv)
#df_resultados_hrv.to_csv("Resultados_hrv.csv", index=False)
```

#### Caractersticas espectrales

```python
def analyze(signal_trans):
    """
    Retorna proporcin del rea, magnitud, picos y promedios totales, as como
    por rangos de frecuencia:
    F1: 0-20 Hz | F2: 20-50 Hz | F3: 50-80 Hz | F4: 80-120 Hz | F5: 120-250 Hz
    """
    N = len(signal_trans)
    frec = 250
    escala = N // frec

    dict = {}

    dict['pico_total'] = np.max(signal_trans)
    dict['promedio_total'] = np.mean(signal_trans)

    F1 = signal_trans[0:20*escala]
    dict['pico_F1'] = np.max(F1)
    dict['promedio_F1'] = np.mean(F1)
    F1_area = simpson(F1)

    F2 = signal_trans[20*escala:50*escala]
    dict['pico_F2'] = np.max(F2)
    dict['promedio_F2'] = np.mean(F2)
    F2_area = simpson(F2)

    F3 = signal_trans[50*escala:80*escala]
    dict['pico_F3'] = np.max(F3)
    dict['promedio_F3'] = np.mean(F3)

    F4 = signal_trans[80*escala:120*escala]
    dict['pico_F4'] = np.max(F4)
    dict['promedio_F4'] = np.mean(F4)

    F5 = signal_trans[120*escala:250*escala]
    dict['pico_F5'] = np.max(F5)
    dict['promedio_F5'] = np.mean(F5)

    dict['area_ratio'] = F2_area / F1_area
    dict['magnitude_ratio'] = dict['pico_F2'] / dict['pico_total']

    return dict
```

```python
fs = 500
base_path = 'ecg_data_denoised/ECGDataDenoised'

def fourier_analysis(urls, fs):
    result = {}
    for url in urls:
        try:
            signal = pd.read_csv(os.path.join(base_path, f"{url}.csv"),
                                 names=['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6'])['II']
        except FileNotFoundError:
            print(f"Archivo no encontrado: {url}.csv")
            continue

        if np.sum(signal) == 0:
            continue

        fft_output = np.fft.fft(signal)
        magnitude_spectrum = np.abs(fft_output)
        magnitude_spectrum = 10 * np.log(magnitude_spectrum)[:2500]

        # Llamar a la funcin analyze (debes asegurarte de definir esta funcin)
        result[url] = analyze(magnitude_spectrum)

    return pd.DataFrame(result)

# Ejecutar el anlisis de Fourier
results_fft = fourier_analysis(df["FileName"], fs=500)

# Combinar resultados FFT con HRV
df_fft = results_fft.T.reset_index().rename(columns={"index": "Registro"})
df_fft_hrv = df_resultados_hrv.merge(df_fft, on="Registro")

# Imprimir los primeros registros combinados
print(df_fft_hrv.head())
```

```python
base_path = 'ecg_data_denoised/ECGDataDenoised'
fs = 500

def analyze_mfcc(mfcc):
    flat = mfcc.flatten()
    return {
        'Promedio': np.mean(flat),
        'Desviacin estndar': np.std(flat),
        'Varianza': np.var(flat),
    }

def compute_mfcc(signal, fs):
    # Asegura que la seal sea flotante y est en el formato correcto
    signal = np.array(signal, dtype=float)
    mfcc = librosa.feature.mfcc(y=signal, sr=fs, n_mfcc=13)
    return mfcc
```

```python
# Verificamos que no haya valores nulos
df_total = df_total.dropna()

# Variables predictoras (todas menos 'Registro' y 'Estado')
X = df_total.drop(columns=["Registro", "Estado"])

# Etiquetas (target)
y = df_total["Estado"]

# Codificamos las clases: AFIB, SR, SB, ST  0, 1, 2, 3
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# One-hot encoding para redes neuronales
y_cat = to_categorical(y_encoded)

# Normalizamos caractersticas
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Divisin entrenamiento/prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_cat, test_size=0.2, random_state=42, stratify=y_cat)
```

###Modelo 1: Red Neuronal Multicapa Simple (MLP bsica)

```python
# Definir modelo
model1 = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(4, activation='softmax')  # 4 clases
])

# Compilar
model1.compile(optimizer=Adam(0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# Entrenar
history1 = model1.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.2, verbose=1)
```

#### Evaluacin del modelo


```python
# Evaluar en test
loss1, acc1 = model1.evaluate(X_test, y_test)
print(f"\nAccuracy del modelo 1 en test: {acc1:.4f}")

# Predicciones
y_pred1 = model1.predict(X_test)
y_pred_classes1 = np.argmax(y_pred1, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Matriz de confusin
cm1 = confusion_matrix(y_true_classes, y_pred_classes1)
plt.figure(figsize=(6, 5))
sns.heatmap(cm1, annot=True, fmt='d', cmap='Blues',
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_)
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.title("Matriz de Confusin - Modelo 1")
plt.show()

# Reporte de clasificacin
print("\nReporte de clasificacin:\n")
print(classification_report(y_true_classes, y_pred_classes1, target_names=label_encoder.classes_))
```

###Modelo 2. Red Neuronal Profunda con Batch Normalization

#### [K - Means](https://www.ibm.com/mx-es/think/topics/k-means-clustering)

La agrupacin de Medias K o K - Means es un algoritmo de aprendizaje no supervisado de clustering iterativo basado en centroides, utilizado para la agrupacin de datos, agrupando puntos de datos no etiquetados en clsteres.

A diferencia del aprendizaje supervisado, los datos de entrenamiento usados no estn etiquetados, por lo que no tienen una estructura de clasificacin definida.

Este algoritmo es de clustering exclusivo o "duro", de tal forma que un punto de datos existe solamente en un clster.

Los grupos se dividen por similitud deacuerdo a la distancia entre sus centroides. El centroide, o centro del clster, es la media o la mediana de todos los puntos dentro del cluster, segn las caractersticas de los datos. Esto surge de minimizar la suma de una medida matemtica de distancia entre los puntos de datos y su centroide asignado, de tal manera que los puntos de datos ms cercanos a un centroide se asisgnan a una misma categora.

```python
df_caracteristicas = pd.read_csv('df_caracteristicas.csv')
df_caracteristicas['Gender'] = df_caracteristicas['Gender'].map({'FEMALE': 1, 'MALE': 0})

# Seleccionar solo las caractersticas (sin nombre de registro ni arritmia)
X = df_caracteristicas.drop(columns=["Registro", "Estado"])


# Escalar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
clusters = kmeans.fit_predict(X_scaled)

# Aadir los clusters al dataframe original
df_caracteristicas["Cluster"] = clusters
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df_caracteristicas["PCA1"] = X_pca[:, 0]
df_caracteristicas["PCA2"] = X_pca[:, 1]

# Comparar clustering con etiquetas reales
plt.figure(figsize=(12, 5))

# Visualizacin por clster asignado
plt.subplot(1, 2, 1)
sns.scatterplot(data=df_caracteristicas, x="PCA1", y="PCA2", hue="Cluster", palette="Set1")
plt.title("Agrupamiento por K-means")

# Visualizacin por arritmia real
plt.subplot(1, 2, 2)
sns.scatterplot(data=df_caracteristicas, x="PCA1", y="PCA2", hue="Estado", palette="Set2")
plt.title("Etiquetas reales de arritmia")

plt.tight_layout()
plt.show()
```
