```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
step_1 = 0.1
step_2 = step_1 / 5

t_1 = np.arange(0, 25 + step_1, step_1)
v_1 = 3 * np.sin(np.pi / 5 * t_1)

t_2 = np.arange(0, 5 + step_2, step_2)
v_2 = 2 * np.sin(np.pi  * t_2)

m_1 = np.vstack((t_1, v_1))
m_2 = np.vstack((t_2, v_2))
```

```python
def analisis(m_1, m_2):
  plt.figure(figsize=(10, 6))

  plt.subplot(2, 1, 1)
  plt.plot(m_1[0], m_1[1], color='green', linewidth=2, linestyle='dashed')
  plt.title('Matriz 1')
  plt.xlabel('Tiempo')
  plt.ylabel('Velocidad')
  plt.xlim(np.min(m_1[0]), np.max(m_1[0]))
  plt.grid(True)
  plt.tight_layout()

  plt.subplot(2, 1, 2)
  plt.plot(m_2[0], m_2[1], color='red', linewidth=2, linestyle='dotted')
  plt.title('Matriz 2')
  plt.xlabel('Tiempo')
  plt.ylabel('Velocidad')
  plt.xlim(np.min(m_2[0]), np.max(m_2[0]))
  plt.grid(True)
  plt.tight_layout()

  plt.show()

  rand_cuerpo1 = m_1[1] + np.random.random(len(m_1[0])) * 20
  rand_cuerpo2 = m_2[1] + (np.random.random(len(m_2[0])) * 15) + 15

  m, b = np.polyfit(rand_cuerpo1, rand_cuerpo2, 1)
  x = np.linspace(np.min(rand_cuerpo1), np.max(rand_cuerpo1), 50)
  y = m*x + b

  plt.figure(figsize=(8, 4))
  plt.scatter(rand_cuerpo1, rand_cuerpo2, color='black', label='Datos')
  plt.plot(x, y, color='red', label=f'Regresión lineal')
  plt.legend()
  plt.title('Regresión lineal')
  plt.xlabel('Cuerpo 1')
  plt.ylabel('Cuerpo 2')
  plt.grid(True)
  plt.tight_layout()

  plt.show()

  return m

```

```python
pendiente = analisis(m_1, m_2)

if pendiente > 0:
    print(f'El movimiento de ambos cuerpos se relaciona: Pendiente={pendiente}')

elif pendiente < 0:
  print(f'El movimiento de ambos cuerpos no se relaciona: Pendiente={pendiente}')

else:
  print(f'La pendiente es cero')
```
